__Импорт и экспорт данных__

3 шага для импорта данных из CSV:

1. Подготовить CSV файл
2. Создать таблицу в базе данных 
3. Выполнить импорт данных из CSV файла в заготовленную таблицу с помощью COPY 

CSV - разновидность текстового файла

Coma Separated Values 

В России чаще всего разделитель - ;

PostgreSQL по умолчанию игнорирует разделители, которые находятся внутри кавычек

PostgreSQL не может сверять имя столбца из файла CSV с названием столбца в таблице баззы данных. Для этого используется HEADER

Если мы хотим импортировать данные из другого файла, то нужно выполнять следующий алгоритм

1. Создать временную таблицу
2. Импортировать туда файлы 
3. Из временно таблицы импортировать в основную с помощью insert into нужные столбцы
4. Удалить временную таблицу
