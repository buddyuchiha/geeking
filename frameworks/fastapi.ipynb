FastAPI - прослойка между фронтендом и базой данных

Мы не можем прямо с фронтенда (прямо с бразуера) запрашивать личные данные, так как поэтому мы их раскроем.

Помимо этого, FastAPI может взаимодействовать с другими API и другими разработчиками через другой язык

Перед FastAPI стоит NGINX (веб-сервер, который позволяет делать много других вещей)

uvicorn позволяет поднимать на наш компьютере небольшой веб сервер

sqlalchemy позволяет работать с базами данных

aiosqlite для взаимодействия с базой SQLite асинхронно

pydantic (устанавливается вместе с fastapi)
