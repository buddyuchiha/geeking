 __Сортировка выбором__

O(n^2)

Есть два указателя i и j, j пробегает по массиву и выбирает самый минимальный элемент, далее эт все обменивается с i

```python
def selection_sort(arr):
  """
  Сортировка выбором (Selection Sort).

  Аргументы:
    arr: Список, который нужно отсортировать.

  Возвращает:
    Отсортированный список.
  """

  n = len(arr)

  # Проходим по всем элементам списка
  for i in range(n):
    # Находим индекс минимального элемента в оставшейся неотсортированной части
    min_index = i
    for j in range(i + 1, n):
      if arr[j] < arr[min_index]:
        min_index = j

    # Меняем местами текущий элемент (arr[i]) с минимальным элементом (arr[min_index])
    arr[i], arr[min_index] = arr[min_index], arr[i]

  return arr


# Пример использования:
my_list = [64, 25, 12, 22, 11]
sorted_list = selection_sort(my_list)
print("Отсортированный список:", sorted_list)  # Вывод: Отсортированный список: [11, 12, 22, 25, 64]


# Другой пример:
another_list = [5, 1, 4, 2, 8]
sorted_another_list = selection_sort(another_list)
print("Отсортированный список:", sorted_another_list) # Вывод: Отсортированный список: [1, 2, 4, 5, 8]

```
